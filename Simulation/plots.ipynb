```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import numpy as np
```

```python
path = os.path.join(os.getcwd(),'data','SINR _ Varying Input Power.csv')
sinr_df = pd.read_csv(path)
```

```python
sinr_df
```

```python
plt.figure(figsize=(10, 6))
# plt.rcParams.update({'font.size': 10})

# for column in sinr_df.columns[5:]:

plt.plot(sinr_df['Factor'].astype(str), sinr_df['DL | snr=100'], marker='X', label='VectorSync',lw='1',ls='-')
# plt.plot(sinr_df['Factor'].astype(str), sinr_df['Neural Network'], marker='o', label='Neural Network',ls='-', lw='1')
plt.plot(sinr_df['Factor'].astype(str), sinr_df['NSB'], marker='D', label='NSB',ls='-', lw='1')
plt.plot(sinr_df['Factor'].astype(str), sinr_df['Capon'], marker='*', label='Capon',ls='-', lw='1')
plt.plot(sinr_df['Factor'].astype(str), sinr_df['Brattel'], marker='P', label='Brattel', ls='-', lw='1')

# Adding labels and title
plt.xlabel('Factor: Times Input UE Power')
plt.ylabel('SINR')
# plt.title('Line Graph of Factors')
plt.legend()
plt.grid(axis='y',ls='--',alpha=0.2)

# Show the plot
plt.show()
```
