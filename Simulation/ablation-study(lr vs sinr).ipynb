```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dl = pd.read_csv(os.path.join(os.getcwd(), 'dl.csv'))
nn = pd.read_csv(os.path.join(os.getcwd(), 'nn.csv'))
nsb = pd.read_csv(os.path.join(os.getcwd(), 'nsb.csv'))
capon = pd.read_csv(os.path.join(os.getcwd(), 'capon.csv'))
bartlett = pd.read_csv(os.path.join(os.getcwd(), 'bartlett.csv'))
```

```python
plt.plot(dl.sinr, dl.factor, label='dl')
plt.plot(nn.sinr, nn.factor, label='nn')
plt.plot(nsb.sinr, nsb.factor, label='nsb')
plt.plot(capon.sinr, capon.factor, label='capon')
plt.plot(bartlett.sinr, bartlett.factor, label='bartlett')

plt.xlabel('sinr')
plt.ylabel('factor')
plt.legend()
plt.show()
```
